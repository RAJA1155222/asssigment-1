Q5. Using while loop, check if A is purely divisible by B and how many times

```python
A = 24
B = 3

count = 0
temp = A

while temp % B == 0:
    temp = temp // B
    count += 1

if count > 0:
    print("A is purely divisible by B")
    print("It is divisible", count, "times")
else:
    print("A is not purely divisible by B")
```

Q6. Create a list of 25 integers and check divisibility by 3

```python
my_list = list(range(1, 26))

for num in my_list:
    if num % 3 == 0:
        print(num, "is divisible by 3")
    else:
        print(num, "is not divisible by 3")
```
